```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('homeprices.csv') #DF - dataframe - Table
```

```python
df
```

```python
plt.scatter(df['area'], df['price'], color='red', marker='+')
plt.xlabel('Area')
plt.ylabel('Price')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
line = LinearRegression()
line.fit(df[['area']], df['price'])
```

```python
# m -> coefficient
line.coef_
```

```python
# c -> intercept
line.intercept_
```

```python
# Predictions
#Predicted_price = m* Area + c
135.78767123*5000 + 180616.43835616432
```

```python
line.predict([[5000]])
```

```python
line.predict([[100]])
```

```python
Area = int(input('Enter Area'))
Predicted_price = line.predict([[Area]])
print(Predicted_price)
```

```python
plt.scatter(df['area'], df['price'], color='red', marker='+')
plt.plot(df['area'], line.predict(df[['area']]), color='blue')
plt.xlabel('Area')
plt.ylabel('Price')
```
